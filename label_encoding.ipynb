```python
def myrange(ret):
    if ret <= -0.02:
        return -2
    elif (ret > -0.02) & (ret < 0):
        return -1
    elif (ret >= 0) & (ret < 0.02):
        return 1
    elif ret >= 0.02:
        return 2
```

```python
df_ret_1d.Close = df_ret_1d.Close.apply(myrange)
```
