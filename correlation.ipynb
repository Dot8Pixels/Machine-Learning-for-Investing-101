```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = yf.download("SPY TDEX.BK GC=F BTC-USD",
                    period="2y", interval="1d")
```

```python
df = data.Close
```

```python
pd.plotting.scatter_matrix(df.pct_change())
```

```python
pd.plotting.scatter_matrix(df.pct_change(), 
                            figsize=(10,10),
                            marker='o',
                            hist_kwds={'bins':20},
                            alpha=0.8)
```

```python
import seaborn as sns
```

```python
sns.pairplot(df.pct_change(), diag_kws={'bins':10})
```

```python
df.corr()
```

```python
df.pct_change().corr()
```

```python
df['BTC-USD'].corr(df['GC=F'])
```

```python
df['BTC-USD'].pct_change().corr(df['GC=F'].pct_change())
```

```python
# plot the heatmap

corr = df.corr()
sns.heatmap(corr,
            xticklabels=corr.columns,
            yticklabels=corr.columns, annot=True)
```
