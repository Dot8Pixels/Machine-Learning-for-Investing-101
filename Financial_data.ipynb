### <b> Resample </b>

```python
df = data.copy()
df.resample('W-Mon').last()
```

```python
df.tail()
```

```python
df_weekly = df.resample('W').agg({'Open' : 'first',
                                'High' : 'max',
                                'Low' : 'min',
                                'Close' : 'last',
                                'Volume' : 'sum'})
```

```python
df_weekly
```

```python
df_weekly = df.resample('W-FRI').agg({'Open' : 'first',
                                'High' : 'max',
                                'Low' : 'min',
                                'Close' : 'last',
                                'Volume' : 'sum'})
```

```python
df_weekly.tail(10)
```

```python
df_monthly = df.resample('BM').agg({'Open' : 'first',
                                'High' : 'max',
                                'Low' : 'min',
                                'Close' : 'last',
                                'Volume' : 'sum'})
```

```python
df_monthly # more information, search keyword is frequency string
```

```python
df.resample('D').last()
```

```python
df_everyday = df.resample('D').last()
```

```python
df_everyday.fillna(method='ffill', inplace=True)
df_everyday
```
