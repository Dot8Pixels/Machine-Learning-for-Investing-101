```python
from sklearn.preprocessing import OneHotEncoder
from sklearn import datasets
import numpy as np
import pandas as pd
```

### Using Sciket-Learn

```python
enc = OneHotEncoder()
```

```python
diabetes = datasets.load_diabetes(as_frame=True)
```

```python
diabetes.data.sex
```

```python
diabetes.data.sex.apply(lambda x: 'male' if x > 0 else 'female')
```

```python
sex = diabetes.data.sex.apply(lambda x: 'male' if x > 0 else 'female')
```

```python
sex
```

```python
enc.fit_transform(sex.values.reshape(-1,1))
```

```python
sex.values.reshape(-1,1)
```

```python
enc.fit_transform(sex.values.reshape(-1,1)).toarray()
```

```python
pd.DataFrame(enc.fit_transform(sex.values.reshape(-1,1)).toarray(), columns=['male', 'female'])
```

```python
test_sex = ['male', 'female', 'LBGT', 'unknown']
```

```python
np.reshape(test_sex,(-1,1))
```

```python
enc.fit_transform(np.reshape(test_sex,(-1,1))).toarray()
```

```python
pd.DataFrame(enc.fit_transform(np.reshape(test_sex,(-1,1))).toarray())
```
