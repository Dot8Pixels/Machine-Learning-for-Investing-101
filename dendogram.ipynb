```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = yf.download('''BTC-USD 
                    ETH-USD 
                    DOGE-USD 
                    ADA-USD
                    SOL1-USD
                    SPY
                    TDEX.BK
                    GC=F''', 
                    period="1y", 
                    interval="1d")
```

```python
df = data.Close
```

```python
df = data.Close.resample("B").last().iloc[2:]
df
```

```python
df = df.fillna(method='ffill')
```

```python
df.head(20)
```

```python
df.tail(20)
```

```python
ret = df.pct_change().iloc[1:]
```

```python
corr = ret.corr()
```

```python
corr
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linked = linkage(corr, 'ward')
labellist = df.columns.to_list()
plt.figure()
dendrogram(linked,
            orientation='top',
            labels=labellist)
plt.show
```

```python
linked = linkage(corr, 'ward')
labellist = df.columns.to_list()
plt.figure()
dendrogram(linked,
            orientation='top')
plt.show
```

```python
corr
```

```python
linked = linkage(corr, 'single')
labellist = df.columns.to_list()
plt.figure()
dendrogram(linked,
            orientation='top',
            labels=labellist)
plt.show
```

```python
linked = linkage(corr, 'centroid')
labellist = df.columns.to_list()
plt.figure()
dendrogram(linked,
            orientation='top',
            labels=labellist)
plt.show
```

```python
linked = linkage(corr, 'ward')
labellist = df.columns.to_list()
plt.figure()
dendrogram(linked,
            orientation='top',
            labels=labellist,
            leaf_rotation=45)
plt.show
```

```python
linked = linkage(corr, 'ward')
labellist = df.columns.to_list()
plt.figure()
dendrogram(linked,
            orientation='left',
            labels=labellist,
            leaf_rotation=0)
plt.show
```
