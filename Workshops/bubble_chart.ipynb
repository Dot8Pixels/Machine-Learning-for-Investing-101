```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('../Example Data/Data_Bubble_Chart.xlsx')
```

```python
df
```

### Method 1

```python
sns.set_style("darkgrid")
plt.rcParams['figure.figsize'] = [9, 7]
```

```python
sns.scatterplot(data=df, x="ROE", y="PBV", size='MARKETCAP', legend=False, hue="Stocks", alpha=0.7, sizes=(200, 2000))

for i, ticker in enumerate (df['Stocks']):
    plt.text(df["ROE"][i]+0.2, df["PBV"][i]+0.2, ticker)
```

### Method 2

```python
fig, ax = plt.subplots()

for i, j in df.iterrows():
    ax.scatter(j.ROE, j.PBV, s = j.MARKETCAP/100, alpha=0.5)
    ax.annotate(j.Stocks, (j.ROE, j.PBV))

plt.title('Bubble Chart')
plt.xlabel('ROE')
plt.ylabel('P/BV')
plt.xlim(0,30)
plt.ylim(0,8)
plt.show()
```
