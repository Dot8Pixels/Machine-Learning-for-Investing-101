```python
import pandas as pd
import yfinance as yf
import seaborn as sns
```

```python
data = yf.download("BBL.BK KBANK.BK BAY.BK TTB.BK KKP.BK TISCO.BK SAWAD.BK MTC.BK AMANAH.BK AEONTS.BK ASK.BK CHAYO.BK JMT.BK	KTC.BK MICRO.BK MTC.BK NCAP.BK S11.BK SAK.BK THANI.BK TK.BK",
                   start = '2018-01-01',end='2021-10-14',period='1d')
```

```python
df = data["Adj Close"].copy()
```

```python
df
```

```python
df.stack().groupby(level=1).head(1)
```

```python
df = df.fillna(method='ffill')
df
```

```python
df_ret = df.pct_change()
```

```python
df_ret = df_ret.loc[~df_ret.isnull().sum(1).astype(bool)]
```

```python
df_ret
```

```python
g = sns.clustermap(df_ret, metric="correlation")
```

```python
df_ret.corr()
```

```python
g = sns.clustermap(df_ret.corr(), method='ward', center = 0.0, cmap="RdYlGn",
                   dendrogram_ratio=(.1, .1),
                   linewidths=.3, figsize=(12, 12), annot = True)
```
