```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../Example Data/workshopdata-labeling.csv')
```

```python
data
```

```python
df = data.copy()
```

```python
df.info()
```

```python
df['Market Cap'] = df['Market Cap'].str.replace(',', '').astype(float)
```

```python
df.info()
```

```python
#Market Cap

df['Label'] = df['Market Cap'].apply(lambda x: 
                3 if x > 1e10 else (
                2 if 2e9 < x < 1e10 else (
                1 if 3e8 < x <2e9 else 0)))
```

```python
df
```

```python
# PE Range

df['PE Range'] = df['PE'].apply(lambda x: 
                'PE<10' if x < 10 else(
                'PE 10-20' if 10< x <20 else(
                'PE 20-50' if 20 < x < 50 else(
                'PE>50' if x > 50 else 'None PE')))) 
```

```python
df
```

```python
# Dividend or Not

df['Dividend'] = ~df['Dividend Yield'].isnull()
```
