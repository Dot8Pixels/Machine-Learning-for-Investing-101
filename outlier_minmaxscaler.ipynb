```python
from sklearn import preprocessing
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
data = fetch_california_housing(as_frame=True)
```

```python
X, y = fetch_california_housing(return_X_y=True)
```

```python
X
```

```python
y
```

```python
transformer = RobustScaler(quantile_range=(10.0, 90.0)).fit(X)
```

```python
X_robust = transformer.transform(X)
```

```python
df_X = pd.DataFrame(X_robust, columns=data.data.columns)
```

```python
df_X
```

```python
df_X.boxplot(figsize=(12, 8))
```

```python
data.data.describe()
```

```python
transformer2 = RobustScaler(quantile_range=(25.0, 75.0)).fit(X)
```

```python
X_robust2 = transformer2.transform(X)
```

```python
pd.DataFrame(X_robust2, columns=df_X.columns).describe()
```

```python
df_X_normalized = pd.DataFrame(X, columns=data.data.columns)
```

```python
df_X_normalized.describe()
```

```python
upper = df_X_normalized.quantile(0.75)
upper
```

```python
low = 0.05
high = 0.95
quant_df = df_X_normalized.quantile([low, high])
```

```python
quant_df
```

```python
filt_df_x = df_X_normalized.apply(lambda x: x[(x >= quant_df.loc[low, x.name]) 
                                    & (x <= quant_df.loc[high, x.name])], 
                                    axis=0)
```

```python
filt_df_x
```

```python
filt_df_x.dropna()
```

```python
Q1 = df_X_normalized.quantile(0.05)
Q3 = df_X_normalized.quantile(0.95)
```

```python
df_X_normalized < Q1
```

```python
df_X_normalized > Q3
```

```python
(df_X_normalized < Q1) | (df_X_normalized > Q3)
```

```python
((df_X_normalized < Q1) | (df_X_normalized > Q3)).any(axis=1)
```

```python
~((df_X_normalized < Q1) | (df_X_normalized > Q3)).any(axis=1)
```

```python
df_X_normalized[~((df_X_normalized < Q1) | (df_X_normalized > Q3)).any(axis=1)]
```

### MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```
