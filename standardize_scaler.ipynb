```python
from sklearn import preprocessing
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
data = fetch_california_housing(as_frame=True)
```

```python
X = data.data
y = data.target
```

```python
df_x = pd.DataFrame(X, columns=data.data.columns)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df_x)
df_x_std = scaler.transform(df_x)
```

```python
df_x_std
```

```python
df_x_std = pd.DataFrame(df_x_std, columns=df_x.columns)
df_x_std
```

```python
df_x2 = df_x.copy()
df_x2.values[:] = StandardScaler().fit_transform(df_x2)
df_x2
```
